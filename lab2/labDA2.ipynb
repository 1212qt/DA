3. Observation of Poisson distribution.

```python
data = {'lambda = 1': np.random.poisson(size=1000, lam=1),
        'lambda = 3': np.random.poisson(size=1000, lam=3),
        'lambda = 10': np.random.poisson(size=1000, lam=10)
       }
df = pd.DataFrame(data)
melt_poiss = df.melt()
g = sns.FacetGrid(melt_poiss, col='variable')
g.map(plt.hist, 'value')
```
